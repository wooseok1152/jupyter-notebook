```python
class Queue:
    def __init__(self):
        self.q = []

    def enQueue(self, item):
        self.q.append(item)

    def deQueue(self):
        if self.isEmpty() == False:
            return self.q.pop(0)

    def size(self):
        return len(self.q)

    def isEmpty(self):
        if len(self.q) > 0:
            return False
        else:
            return True

    def peek(self):
        if self.isEmpty() == False:
            return self.q[0]

    def delete(self, item):
        if item in self.q:
            self.q.remove(item)
        else:
            print("해당 아이템이 존재하지 않습니다.")

import numpy as np

cur_time = 0  # 현재 시각


class Customer:
    def __init__(self, time, Queue):
        # 손님의 음료가 만들어지는데 소요되는 시간
        self.cook_time = np.random.normal(1, 0.2)

        # 손님들이 가게에 평균적으로 1분마다(시간) 입장함, random.exponential의 첫번째 매개변수는 '평균시간'을 의미한다.
        self.term = np.random.exponential(1)

        # 손님이 가게에 도착한 시각
        self.arrive_time = cur_time + self.term

        # 손님이 음료를 주문하는 시각

        # 대기손님이 아무도 없을 때
        if len(Queue.q) == 0:
            self.oder_time = self.arrive_time

        elif self.arrive_time > Queue.q[-1].out_time:
            self.oder_time = self.arrive_time

        # 새로운 손님이 도착했는데 이미 다른 손님들이 줄을 기다리고 있을 때
        else:
            self.oder_time = Queue.q[-1].out_time

        # 손님이 나가는 시각(self.oder_time에 이미 해당 손님 도착 시간이 내포되어 있음)
        self.out_time = self.oder_time + self.cook_time


# Shop(매장) 클래스 정의
class Shop:

    def __init__(self):
        self.customer_queue_line = Queue()

    def line_size(self):
        return self.customer_queue_line.size()

    def customer_enter(self, customers):
        return self.customer_queue_line.enQueue(customers)

    def customer_out(self):
        return self.customer_queue_line.deQueue()

    def get_last_customer(self):
        if self.customer_queue_line.size() != 0:
            return self.customer_queue_line.q[-1]


choi_coffe_shop = Shop()

while cur_time < 720:

    i = 0
    # 'customer'라는 객체가 반복적으로 생성되고 해당 객체의 id는 전부 다르다,
    customer = Customer(cur_time, choi_coffe_shop.customer_queue_line)
    choi_coffe_shop.customer_enter(customer)
    cur_time = customer.arrive_time

    while True:
        if choi_coffe_shop.customer_queue_line.q[0].out_time < cur_time:
            choi_coffe_shop.customer_out()
        else:
            i += 1
            if i > 0:
                break
        print('%.2f, %.2f, %.2f, %.2f' % (cur_time, customer.arrive_time, customer.oder_time, customer.out_time),
      choi_coffe_shop.line_size())
                
            
     




```
