```python
import requests
from bs4 import BeautifulSoup
import urllib.request

#에픽세븐 이미지 다운
req = requests.get('https://epicsevendb.com/heroes')
html = req.text

soup = BeautifulSoup(html, 'html.parser')

names = soup.select('.columns > li')

hero_name = []
img_url = []

for i in names:
    hero_name.append(i['id'])

for i in hero_name:
    img_url.append("https://assets.epicsevendb.com/herofull/" + i + ".png")

num = 0

for i in img_url:
    num += 1
    name = hero_name[num] + ".png"
    try:
        opener = urllib.request.build_opener()
        opener.addheaders = [('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
        urllib.request.install_opener(opener)
        urllib.request.urlretrieve(i, "./epic/"+str(name)+'.png')
    except:
        print("\n")
```

```python
#언리쉬드 이미지 다운
for i in range(1, 300):
    try:
        num = 10000000 + i
        url = 'http://uldb.iustice.net/wiki/NoxDetail.aspx?cd_nox=' + str(num)
        base_url = "http://uldb.iustice.net/wiki/"

        req = urllib.request.Request(url)
        res = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(res,'html.parser')
        soup = soup.find_all("img")[1]["src"]

        img_url = base_url + soup

        urllib.request.urlretrieve(img_url, "./unleashed/"+str(i)+'.jpg')
    except:
        print("\n")
```

```python
#소녀전선 이미지 다운

import ssl

ssl._create_default_https_context = ssl._create_unverified_context

for i in range(1,300):
    try:
        context = ssl._create_unverified_context()
        url = "https://gfl.zzzzz.kr/doll.php?id="+str(i)+"&lang=ko"
        base_url = "https://gfl.zzzzz.kr/"
        req = urllib.request.Request(url)
        res = urllib.request.urlopen(req, context=context).read()

        soup = BeautifulSoup(res,'html.parser')
        soup = soup.select(".doll_img > img")[0]['src']

        img_url = base_url + soup

        urllib.request.urlretrieve(img_url, "./shojo/"+str(i)+'.jpg')
    except:
        print("\n")
```

```python
#소녀전선 중상 이미지 다운

import ssl

ssl._create_default_https_context = ssl._create_unverified_context

for i in range(1,300):
    try:
        context = ssl._create_unverified_context()
        url = "https://gfl.zzzzz.kr/doll.php?id="+str(i)+"&lang=ko"
        base_url = "https://gfl.zzzzz.kr/"
        req = urllib.request.Request(url)
        res = urllib.request.urlopen(req, context=context).read()

        soup = BeautifulSoup(res,'html.parser')
        soup = soup.select(".doll_img > img")[0]['src']
        soup = soup.replace(".png", "_d.png")

        img_url = base_url + soup

        urllib.request.urlretrieve(img_url, "./shojo_d/"+str(i)+'.jpg')
    except:
        print("\n")
```

```python
#라스트오리진 이미지 다운

url = "https://lastorigin.fandom.com/wiki/Unit_List"
base_url = "https://lastorigin.fandom.com/"
req = urllib.request.Request(url)
res = urllib.request.urlopen(req).read()
soup = BeautifulSoup(res,'html.parser')
soup = soup.select(".image")

char_name = []

for i in range(len(soup)):
    char_name.append(soup[i]["href"])
    char_name[i] =  base_url + char_name[i]

num = 0

for i in char_name:
    num += 1
    url = i
    req = urllib.request.Request(url)
    res = urllib.request.urlopen(req).read()
    soup = BeautifulSoup(res,'html.parser')
    img_url = soup.select(".pi-item > a")[0]["href"]
    
    urllib.request.urlretrieve(img_url, "./last/"+str(num)+'.jpg')
```
