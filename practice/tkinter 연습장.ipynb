```python
import tkinter as tk
from tkinter import filedialog
import pandas as pd
from PIL import ImageTk, Image

photo = ImageTk.PhotoImage(Image.open(r"C:\Users\my\Desktop\test.jpg"))

def address():
    
    window.dirName = filedialog.askopenfilename(initialdir = "C:/", title = "엑셀파일 선택", filetypes = (("xlsx files","*.xlsx"),("all files","*.*")))
    print(window.dirName)
    
def process_excel():
    
    result_label.config(text = "complete!", anchor = "center")
    sheetName = entry1.get()
    attributeName = entry2.get()
    word = entry3.get()
    print(sheetName, attributeName, word)
    
window = tk.Tk()
window.title ("For Lovelyn")
window.geometry("800x500+%d+150" % (660))

leftFrame = tk.Frame(window, relief = "solid", bd = 1)
leftFrame.pack(side = "left", anchor = "n", pady = 5)

frame1 = tk.Frame(leftFrame, width = 50)
frame1.pack(anchor = "w")

file_button = tk.Button(frame1, text = "파일 선택", width = 27, command = address)
file_button.pack(anchor = "center")

frame2 = tk.Frame(leftFrame)
frame2.pack(anchor = "w")

sheet_label = tk.Label(frame2, text = "시트명 : ")
sheet_label.grid(row = 0, column = 0)

entry1 = tkinter.Entry(frame2)
entry1.grid(row = 0, column = 1)

frame3 = tk.Frame(leftFrame)
frame3.pack(anchor = "w")

attribute_label = tk.Label(frame3, text = "변수명 : ")
attribute_label.grid(row = 0, column = 0)

entry2 = tkinter.Entry(frame3)
entry2.grid(row = 0, column = 1)

frame4 = tk.Frame(leftFrame)
frame4.pack(anchor = "w")

word_label = tk.Label(frame4, text = " 단어 :   ")
word_label.grid(row = 0, column = 0)

entry3 = tkinter.Entry(frame4)
entry3.grid(row = 0, column = 1)

frame4 = tk.Frame(leftFrame)
frame4.pack(anchor = "w")

file_button = tk.Button(frame4, text = "클릭!", width = 27, command = process_excel)
file_button.pack(anchor = "center")

result_label = tk.Label(leftFrame, width = 27)
result_label.pack(anchor = "w")


rightLabel = tk.Label(window, relief = "solid", width = 200, height = 500, image = photo)
rightLabel.pack(side = "left", anchor = "n", padx = 5, pady = 5)

window.mainloop()
```
