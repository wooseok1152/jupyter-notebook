```python
# 같은 성별끼리 가격을 합하고, 해당 합에 대한 열을 추가하기

ex = pd.DataFrame({"성별" : ["남" , "남", "여", "여"], "가격" : [10, 20, 30, 40]})
list_sex = ex["성별"]
list_price = ex["가격"]
list_price_sum = []

print("list_sex :")
print(list_sex, "\n")

print("list_price :")
print(list_price, "\n")

for i in range(len(list_sex)):
    
    for j in range(len(list_sex)):
        
        if i == j:
            
            pass
        
        else:
            
            if list_sex[i] == list_sex[j]:

                list_price_sum.append(list_price[i] + list_price[j])

ex["총합"] = list_price_sum
print("result :")
print(ex)
```
