```python
def max(x):
    max_index = 0
    max_value = x[0]
    for i in range(len(x)):
        if x[i] > max_value:
            max_value = x[i]
            max_index = i
    return max_index

def sort(x):
    for i in range(len(x)):
        _tmp = x[i:]
        maxindex_in_tmp = max(_tmp)
        x[i], x[maxindex_in_tmp + i] = x[maxindex_in_tmp + i], x[i] 
    return x

def bigger_price(limit: int, data: list) -> list:
    all_price_list = []

    for i in range(len(data)):
        all_price_list.append(data[i]["price"])
    print(all_price_list)

    sort(all_price_list)
    
    print(all_price_list)

    sorted_data = []

    for i in range(len(data)):
        for j in range(len(data)):
            if data[j]["price"] == all_price_list[i]:
                sorted_data.append(data[j])
    print("sorted_data is ", sorted_data)

    final_result_data = []

    for i in range(0, limit):
        final_result_data.append(sorted_data[i])
    
    print("final_result_data =", final_result_data)

    return final_result_data


if __name__ == '__main__':
    from pprint import pprint
    print('Example:')
    pprint(bigger_price(2, [
        {"name": "bread", "price": 100},
        {"name": "wine", "price": 138},
        {"name": "meat", "price": 15},
        {"name": "water", "price": 1}
    ]))

    # These "asserts" using for self-checking and not for auto-testing
    assert bigger_price(2, [
        {"name": "bread", "price": 100},
        {"name": "wine", "price": 138},
        {"name": "meat", "price": 15},
        {"name": "water", "price": 1}
    ]) == [
        {"name": "wine", "price": 138},
        {"name": "bread", "price": 100}
    ], "First"

    assert bigger_price(1, [
        {"name": "pen", "price": 5},
        {"name": "whiteboard", "price": 170}
    ]) == [{"name": "whiteboard", "price": 170}], "Second"

    print('Done! Looks like it is fine. Go and check it')
```
