```
G = nx.Graph()
relations = [('apple','banana'), ('kiwi','apple'),('mango','apple')]
G.add_edges_from(relations)
nx.draw(G)
```

```
pos = nx.spectral_layout(G)
print(pos)
```

```
print(G.degree())
```

```
degree = G.degree()
nx.draw(G, pos, node_size = [v*1000 for _, v in degree])
nx.draw_networkx_labels(G, pos)
```

```
nx.to_numpy_matrix(G)
```

```
print(nx.betweenness_centrality(G))
print(nx.closeness_centrality(G))
print(nx.degree_centrality(G))
print(nx.eigenvector_centrality(G))
```

```
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_node("ANUJ")
G.add_node("BOB")
G.add_node("PEGGY")
G.add_node("YOU")
G.add_node("ALICE")
G.add_node("CLAIRE")
G.add_node("JONNY")
G.add_node("THOM")
G.add_edge('ANUJ','BOB')
G.add_edge('BOB','PEGGY')
G.add_edge('BOB','YOU')
G.add_edge('ALICE','PEGGY')
G.add_edge('ALICE','YOU')
G.add_edge('YOU','CLAIRE')
G.add_edge('CLAIRE','THOM')
G.add_edge('CLAIRE','JONNY')
```

```
pos_spring = nx.spring_layout(G)
pos_fr = nx.fruchterman_reingold_layout(G)
pos_fr
nx.draw(G, pos=pos_fr, node_size=3000, node_color='red', with_labels=True)
```

```
from operator import itemgetter
eig = nx.eigenvector_centrality(G)
# let's list the scores
sorted(eig.items(), key=itemgetter(1), reverse=True)
```

```
degree = G.degree()
degree = list(degree)
degree
```

```
nx.draw(G, pos=pos_fr, node_color='red', with_labels=True, node_size = [v*1000 for _, v in degree])
```

```
list(eig.items())
```

```
nx.draw(G, pos=pos_fr, node_color='red', with_labels=True, node_size = [v*1000 for _, v in list(eig.items())])
```
