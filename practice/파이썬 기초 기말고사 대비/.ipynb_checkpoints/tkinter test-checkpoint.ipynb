```python
# label에 표시되어있는 내용을 button과 entry를 이용하여 바꾸기

import tkinter as tk

count = 0


class window:
    def __init__(self):
        window = tk.Tk()
        window.title("최우석입니다.")
        window.geometry("800x500")
        window.resizable(False, False)
        
        self.label1 = tk.Label(window, text = "안녕하세요.", width = 50, height = 20, relief = "solid", fg = "red")
        self.label1.pack(pady = 10)
        
        
        # button은 event에 대한 처리를 'command'매개변수로 한다. chenkbutton과 radiobutton도 마찬가지이다.
        # 이외 다른 위젯(ex : listbox)은 'bind' 메쏘드를 이용하여 event를 처리한다.
        self.button1 = tk.Button(window, text = "클릭하세요", width = 10, height = 5, overrelief = "solid", command = self.countup)
        self.button1.pack(pady = 10)
        
        self.entry1 = tk.Entry(window)
        
        # entry를 작성한 후 엔터를 누르면("Retunr"의 의미), 'self.number_show'메소드(window 클래스의 메소드)가 실행된다.
        self.entry1.bind("<Return>", self.number_show)
        
        self.entry1.pack(pady = 10)
        
        # 'window.mainloop()'는 생성자 함수 정의 블락에서 항상 맨 밑에 적혀야한다.
        window.mainloop()
        
    # bind를 통해 실행되는 함수는 매개변수로 무조건 event를 가져야 한다.
    def number_show(self, event):
        self.label1.config(text = str(self.entry1.get()))
        
    def countup(self):
        
        # 전역변수를 함수 내에서 업데이트 시킬 때는 global 키워드를 사용해야 한다.
        global count
        
        count = count + 1
        self.label1.config(text = str(count))
        
        
a = window()
```

```python
# pad를 이용하여 두 개의 button을 pack으로 배치하기

import tkinter as tk

class Window:
    def __init__(self):
        window = tk.Tk()
        window.geometry("640x400")
        window.resizable(False, False)
        b1 = tk.Button(window, text="Python first Button", height=5, width=30)
        b2 = tk.Button(window, text="Python second Button", height=5, width=30)
        
        # pack의 매개변수 padx와 pady는 해당 위젯 기준 왼쪽, 오른쪽 또는 위, 아래에 전부 여백을 둔다.
        b1.pack(side="left", padx=10)
        b2.pack(side="left", padx=0)
        window.mainloop()

a = Window()
```

```python
# 리스트 박스 활용하기

import tkinter

window=tkinter.Tk()
window.title("YUN DAE HEE")
window.geometry("640x480+100+100")
window.resizable(False, False)

# 'selectmode' 매개변수는 요소들의 선택 형식을 뜻한다. 
#'extended'는 화살표로 요소들을 선택할 수 있고, 여러 요소들을 shift 키와 마우스 드래그를 통해 선택할 수 있다.
listbox = tkinter.Listbox(window, selectmode='extended', height=0)

listbox.insert(0, "1번")
listbox.insert(1, "2번")
listbox.insert(2, "2번")
listbox.insert(3, "2번")
listbox.insert(4, "3번")
listbox.delete(1, 2)
listbox.pack()

window.mainloop()
```

```python
# label 2개, entry 1개, button 1개를 활용하여 입력한 숫자가 홀수인지 짝수인지 보여주는 프로그램 만들기

import tkinter as tk

class Window:
    def __init__(self):
        window = tk.Tk()


        label = tk.Label(window, text="정수를 입력하세요", width=15, height=5)
        self.entry = tk.Entry(window, width=11)
        button = tk.Button(window, text="홀수/짝수?", command=self.clickEvent)
        label.pack(side="left", padx=4)
        self.entry.pack(side="left")
        button.pack(side="left", padx=4)
        
        # 홀수인지 짝수인지 알려주는 label은 초기에 아무것도 안타나져있다.
        self.result = tk.Label(window, text="", width=15, height=5)
        
        self.result.pack()
        window.mainloop()

    def clickEvent(self):
        
        # entry의 get 메소드는 박스안의 문자를 string으로 가져온다.
        x = int(self.entry.get())
        
        if x % 2 == 0:
            self.result.config(text="짝수입니다.")
        else:
            self.result.config(text="홀수입니다.")

a = Window()
```

```python
# checkbutton 위젯 활용

import tkinter as tk

class Window:
    def __init__(self):
        window=tk.Tk()
        window.geometry("640x480")
        window.resizable(False, False)
        self.CheckVariety_1 = tk.IntVar()
        self.CheckVariety_2 = tk.IntVar()
        self.CheckVariety_3 = tk.IntVar()
        self.CheckVariety_4 = tk.IntVar()


        # 특정 checkbutton이 눌러지면 value = 1이 되고, 안눌러지면 value = 0이 된다. 그래서 value 매개변수가 굳이 없다.
        checkbutton1 = tk.Checkbutton(window, text="O", variable=self.CheckVariety_1, activebackground="blue", command=self.selected)
        checkbutton2 = tk.Checkbutton(window, text="△", variable=self.CheckVariety_2, command=self.selected)
        checkbutton3 = tk.Checkbutton(window, text="X", variable=self.CheckVariety_3, command=self.selected)
        checkbutton4 = tk.Checkbutton(window, text="xx", variable=self.CheckVariety_4, command=self.selected)

        checkbutton1.pack()
        checkbutton2.pack()
        checkbutton3.pack()
        checkbutton4.pack()

        window.mainloop()

    def selected(self):
        print(self.CheckVariety_1.get())
        print(self.CheckVariety_2.get())
        print(self.CheckVariety_3.get())
        print(self.CheckVariety_4.get())

a = Window()
```

```python
# weather 나타내는 프로그램

import tkinter as tk
from urllib.request import urlopen
import urllib.request, json
from PIL import ImageTk,Image


class Window:
    def __init__(self):
        self.url1 = "http://api.openweathermap.org/data/2.5/weather?id="
        self.url2 = "&APPID=2ae88b3ec4b4537d15165fca377a3ca9"
        self.window = tk.Tk()
        self.window.geometry("640x480")
        self.window.resizable(False, False)
        self.window.title("Weather Pro")

        self.f1 = tk.Frame(self.window, bd=2, padx=5, pady=5)
        self.f2 = tk.Frame(self.window, bd=2, padx=5, pady=5)

        file2 = open(r"C:\Users\user\Desktop\weatherPro\city.list.json", encoding='utf-8-sig').read()
        cities = json.loads(file2)

        listbox = tk.Listbox(self.f1, selectmode='extended', height=0)

        self.nameID = {}
        for city in cities:
            if city['country'] == 'KR':
                self.nameID[city['name']] = city['id']

        cityList  = ['Yeoju','Yanggu','Mungyeong','Kurye','Chinchon','Hamyang','yeoncheongun','Yangpyong','Yangju','Wanju','Waegwan','Vijongbu','Puan','Asan','Munsan','Kwangju','Kuri','Kumi','Gimcheon']
        cityList += ['Gapyeong','Hwaseong','Hwacheon','Hongsung','Hongchon','jin-angun','Tonghae','Seongnam','Hwado','Changnyeong','Taian','Songwon','Kosong','Kijang','Imsil','Sinan','Andong']
        cityList += ['Ulsan','Busan','Ansan','Daegu','Kimhae','Munemi','Wonju','Masan','Iksan','Yongin','Pyongtak','Chinju','Bucheon','Gwangju','Jeju','Yangsan','Cheongju','Chuncheon','Goyang']
        cityList += ['Yanggok','Tenan','Yesan','Kongju','Chinhae','Kunsan','Moppo','Anyang','Jeonju','Reisui','Sogcho','Hanam','Kang-neung','Sunchun','Miryang','Hadong','Sangju']
        cityList += ['Suisan','Osan','Ichon','Kimje','Nonsan','Inje','Yangyang','Goseong','Muju','Nangen','Haenam','Muan','Hwasun','Naju','Kyonju','Namhae','Kwangmyong','Kochang','Taebaekv']
        cityList += ['Koesan','Pyeong','Ulchin','Namyang','Tangjin','Kanghwa','Daejeon','Suigen','Seoul','Incheon','Changwon','Anseong','Wabu']
        cityList.sort()
        i = 0
        for city in cityList:
            listbox.insert(i,city)
            i += 1
            
        #listbox는 event 처리를 'bind'메쏘드로 한다.
        listbox.bind("<<ListboxSelect>>", self.selected)
        listbox.pack()
        self.imgLabel = tk.Label(self.f2, width=210, height=150, relief=tk.RAISED, pady=50)
        self.imgLabel.pack()
        img = Image.open(r"C:\Users\user\Desktop\weatherPro\weatherIcon\open1.jpg")
        image = ImageTk.PhotoImage(img)
        self.imgLabel.config(image=image)
        img.close()

        self.label1 = tk.Label(self.f2, text="날씨: " , width=30, height=5, fg="red", relief=tk.RAISED, pady=25)
        self.label2 = tk.Label(self.f2, text="온도: " , width=30, height=5, fg="red", relief=tk.RAISED, pady=25)
        self.label3 = tk.Label(self.f2, text="습도: " , width=30, height=5, fg="red", relief=tk.RAISED, pady=25)
        self.label1.pack()
        self.label2.pack()
        self.label3.pack()
        self.f1.pack(side="left")
        self.f2.pack()
        self.window.mainloop()

    def selected(self,event):
        w = event.widget
        index = int(w.curselection()[0])
        value = w.get(index)
        url = self.url1 + str(self.nameID[value]) + self.url2
        file1 = urllib.request.urlopen(url)
        s = json.loads(file1.read())
        # print(s)
        # print(s['weather'][0]['description'], s['weather'][0]['icon'])
        # print(s['main']['temp'] - 273.15, s['main']['humidity'] )
        iconURL = 'http://openweathermap.org/img/w/' + s['weather'][0]['icon'] +'.png'
        image = Image.open(urlopen(iconURL))

        if image.size[1] > image.size[0]:
            hSize = int((200 * image.size[0] / image.size[1]))
            vSize = 200
        else:
            hSize = 200
            vSize = int((200 * image.size[1] / image.size[0]))

        image = image.resize((hSize,vSize), Image.ANTIALIAS)
        _image = ImageTk.PhotoImage(image)
        self.imgLabel.config(image=_image)
        image.close()
        self.label1.configure(text = "날씨: " + s['weather'][0]['description'])
        self.label2.configure(text = "온도: " + str(int(s['main']['temp'] - 273.15)))
        self.label3.configure(text = "습도: " + str(s['main']['humidity']))
        self.window.mainloop()

a = Window()

```
