```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sig(x):
    return 1/(1+np.exp(-x))

def deriv(w):
    sum = 0
    for i in range(n):
        sum += sig(w*x[i])*x[i] - y[i]
    return sum

def cost(w):
    sum = 0
    for i in range(n):
        temp1 = -y[i] * np.log(sig(w*x[i]))
        temp2 = -(1-y[i]) * np.log(1-sig(w*x[i]))
        #print(w, temp1, temp2)                       
        sum += temp1+temp2
        return sum


w = 0; b= -10

x = np.array([1,2,3,4,5,6])
y = np.array([0,0,0,1,1,1])
xy = np.array([[1,0],[2,0],[3,0],[4,1],[5,1],[6,1]])
n = y.shape

n = 100
a = np.linspace(-5, 5, n)
b = np.linspace(-2.5, 2.5, n)
a, b = np.meshgrid(a, b)

```

```python
cost1 = np.zeros((100, 100))
for val in xy:
    temp = -val[1] * np.log(sig(a+b*val[0])) - (1-val[1]) * np.log(1-sig(a+b*val[0]))
    cost1 += temp

cost2 = np.zeros((100, 100))
for val in xy:
    temp = (val[1]-sig(a+b*val[0]))**2
    cost2 += temp
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(a, b, cost1, 100, cmap='binary')
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(a, b, cost2, 100, cmap='binary')
```
