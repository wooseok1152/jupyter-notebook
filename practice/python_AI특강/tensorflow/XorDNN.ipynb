```
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
import time
import tensorflow as tf
import numpy as np

x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

#h1: 10 nodes, h2: 7 nodes
# nh1=10
# nh2=7
# W1 = tf.Variable(tf.random_uniform([2, nh1], -1.0, 1.0), name='weight1')
# W2 = tf.Variable(tf.random_uniform([nh1, nh2], -1.0, 1.0), name='weight2')
# W3 = tf.Variable(tf.random_uniform([nh2, 1], -1.0, 1.0), name='weight3')

# b1 = tf.Variable(tf.zeros([nh1]), name='bias1')
# b2 = tf.Variable(tf.zeros([nh2]), name='bias2')
# b3 = tf.Variable(tf.zeros([1]), name='bias3')

# h1 = tf.sigmoid(tf.matmul(X, W1)+b1)
# h2 = tf.sigmoid(tf.matmul(h1, W2)+b2)

# yhat = tf.sigmoid(tf.matmul(h2, W3)+b3)

nh=10
W1 = tf.Variable(tf.random_uniform([2, nh], -1.0, 1.0), name='weight1')
W2 = tf.Variable(tf.random_uniform([nh, nh], -1.0, 1.0), name='weight2')
W3 = tf.Variable(tf.random_uniform([nh, nh], -1.0, 1.0), name='weight3')
W4 = tf.Variable(tf.random_uniform([nh, nh], -1.0, 1.0), name='weight4')
W5 = tf.Variable(tf.random_uniform([nh, nh], -1.0, 1.0), name='weight5')
W6 = tf.Variable(tf.random_uniform([nh, nh], -1.0, 1.0), name='weight6')
W7 = tf.Variable(tf.random_uniform([nh, 1], -1.0, 1.0), name='weight7')
b1 = tf.Variable(tf.zeros([nh]), name='bias1')
b2 = tf.Variable(tf.zeros([nh]), name='bias2')
b3 = tf.Variable(tf.zeros([nh]), name='bias3')
b4 = tf.Variable(tf.zeros([nh]), name='bias4')
b5 = tf.Variable(tf.zeros([nh]), name='bias5')
b6 = tf.Variable(tf.zeros([nh]), name='bias6')
b7 = tf.Variable(tf.zeros([1]), name='bias7')
h1 = tf.nn.relu(tf.matmul(X, W1)+b1)
h2 = tf.nn.relu(tf.matmul(h1, W2)+b2)
h3 = tf.nn.relu(tf.matmul(h2, W3)+b3)
h4 = tf.nn.relu(tf.matmul(h3, W4)+b4)
h5 = tf.nn.relu(tf.matmul(h4, W5)+b5)
h6 = tf.nn.relu(tf.matmul(h5, W6)+b6)
yhat = tf.sigmoid(tf.matmul(h6, W7)+b7)



# cost/loss function
cost = -tf.reduce_mean(Y * tf.log(yhat) + (1 - Y) * tf.log(1 - yhat))

train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(cost)

# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(yhat > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype=tf.float32))

# Launch graph

start=time.time()
sess = tf.Session()

# Initialize TensorFlow variables
sess.run(tf.global_variables_initializer())

for step in range(10001):
    sess.run(train, feed_dict={X: x_data, Y: y_data})
    if step % 1000 == 0:
        print(step, sess.run(cost, feed_dict={X: x_data, Y: y_data}))

end = time.time() - start
print(end)

# Accuracy report
h, c, a = sess.run([yhat, predicted, accuracy], feed_dict={X: x_data, Y: y_data})
print(h)
```
