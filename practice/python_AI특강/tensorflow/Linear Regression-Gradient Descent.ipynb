```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

```python
n = 100
a = np.linspace(-2, 2, n)
b = np.linspace(-1, 3, n)
a, b = np.meshgrid(a, b)
```

```python
a.shape, b.shape
```

```python
xy = [[1,1],
      [2,2],
      [3,3]]

cost =np.zeros((100, 100))
for val in xy:
    temp = (val[1]- a - b*val[0]) * (val[1]- a - b*val[0])
    cost += temp
```

```python
cost.shape
```

```python
ax = plt.axes(projection='3d')
ax.contour3D(a, b, cost, 100, cmap='binary')
```
