```
import numpy as np

x1 = [0,0,1,1]
x2 = [0,1,0,1]
w1 = 9
w2 = 9
b = -4
# w1 = -0.16
# w2 = 0.44
# b = -0.99
for i in range(4):
    summ = w1*x1[i] + w2*x2[i] + b
    print(1/(1+np.exp(-summ)))
```
