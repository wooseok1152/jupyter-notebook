### <font color='blue'>파이썬의 특징 </font>
* 다양한 외부 패키지가 존재함

    프로그램을 잘 하지 못하는 사람도 파이썬을 쉽게 사용할 수 있음

### <font color='blue'> Passing by value </font> 

```
def try_to_modify(a, b, c):
    a = 23
    b.append(42)
    c = [99]
    print(a, b, c)
    print(id(b))
    print(id(c))
```

```
# If the value is mutable, the function may modify the caller’s variable in-place:

a = 77     # immutable variable
b = [99]   # mutable variable
c = [28]
try_to_modify(a, b, c)
print(a, b, c)
print(id(b))
print(id(c))
```

### <font color='blue'> OOP(Object Oriented Programming) </font>  

```
class Student(object):
    id = 0 # statis variable
    def __init__(self, name):
        Student.id += 1
        self.name = name
    def setAge(self, age):
        self.age = age
    def setMajor(self, major):
        self.major = major
```

```
anna = Student('anna')
anna.setAge(21)
anna.setMajor('Physics')
print(anna.id, Student.id)
```

```
john = Student('john')
john.setAge(23)
john.setMajor('Statistics')
print(john.id, john.major)
```

```
# The MasterStudent class inherited from the Student attributes and methods
class MasterStudent(Student):
    def setDate(self, endDate):
        self.endDate = endDate

james = MasterStudent('james')
james.setDate('June')
james.setAge(23)
print(james.age, james.endDate)
```

## <font color='blue'> pandas </font>  

* pandas는 대규모 데이터 셋을 처리하는 패키지이다.
* 판다스는 일반적으로 행은 개체를 나타내고 열은 변수를 나타낸다.
* 아래는 통계학의 유명한 데이터셋 iris dataset을 판다스 데이터 프레임으로 가져오는 예이다.
* iris 데이터는 붓꽃의 종류(setosa, versicolor, virginica)에 따른 붓꽃의 꽃잎 길이와 폭, 꽃순의 길이와 폭을 정리한 자료이다.

```
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
print(data.keys())
```

```
print(data['target_names'])
```

```
print(data['DESCR'])
```

```
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```
len(df)
```

```
df['species'] = data.target
df
```

```
X = df.as_matrix(columns=df.columns[1:4])
y = df.as_matrix(columns=df.columns[4:])
X, y
```

```
import matplotlib
import matplotlib.pyplot as plt
sl = df.iloc[:,0]
sw = df.iloc[:,1]
species = df.iloc[:,4]

```

```
print(species)
```

```
cmap = matplotlib.cm.get_cmap('viridis')
normalize = matplotlib.colors.Normalize(vmin=min(species), vmax=max(species))
colors = [cmap(normalize(value)) for value in species]
plt.scatter(sl, sw, color=colors)
plt.show()
```

```
from sklearn.preprocessing import scale, minmax_scale
Z = minmax_scale(X)
```
