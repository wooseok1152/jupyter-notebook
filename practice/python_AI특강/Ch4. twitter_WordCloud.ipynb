## 4-1크롤링(Crawling) 이란?

* 크롤링은 인터넷 상의 컴퓨터 서버에서 원하는 내용을 가져오는 것을 의미한다.
* 일반적으로 네이버, 트위터, 페이스북 등 많은 기업들이 API 라는 방식의 정보제공 서비스를 하고 있다.
* 이를 이용하기 위해서는 API를 사용할 수 있는 권한을 해당 회사로 부터 획득하여야 한다.
* 권한 없이 특정 서버의 데이터를 가져오는 것이 해킹이다.

```python
search = [] # 크롤링 결과 저장할 변수
cnt = 1
while(cnt <= 100):   # 10page 대상으로 크롤링
    tweets = api.search(keyword)
    for tweet in tweets:
        if tweet.text.find("RT ") == -1 :search.append(tweet.text)
    cnt += 1
```

```python
from konlpy.tag import Kkma
from konlpy.utils import pprint
kkma = Kkma()
pprint(kkma.sentences(u'네, 안녕하세요. 반갑습니다.'))
```

```python
nouns = kkma.nouns(u'질문이나 건의사항은 깃헙 이슈 트래커에 남겨주세요.')
```

```python
nouns = []
for i in range(len(search)):
    temp = [s for s in kkma.nouns(search[i]) if "파이썬" not in s]
    temp = [s for s in temp if "python" not in s]
    temp = [s for s in temp if len(s) >= 2]
    nouns.append(temp)
```

```python
import nltk
noun_freq = nltk.Text(nouns, name = "파이썬")
data = noun_freq.vocab().most_common(100)
```

```python
!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

tmp_data = dict(data)
wordcloud = WordCloud(font_path='AppleGothic.ttf', background_color='white').generate_from_frequencies(tmp_data)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
