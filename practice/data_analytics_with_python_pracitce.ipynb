```python
# cost 열 내 원소들을 int 자료형으로 변환시키고, 해당 열 내 원소들의 합과 평균을 계산

df = pd.read_csv(r"C:\Users\my\Desktop\supplier_data.csv")
cost_vector = df["Cost"]
new_cost_vector = []
for i in range(len(cost_vector)):
    
    cost_vector[i] = cost_vector[i][1:]
    str_to_list = list(cost_vector[i])
    for j in range(len(str_to_list)):
        
        if str_to_list[j] == ",":
        
            str_to_list[j] = ""
            
    list_to_str = "".join(str_to_list)
    list_to_str = int(float(list_to_str))
    new_cost_vector.append(list_to_str)
    
# print(new_cost_vector, "\n")
    
df["new_cost"] = new_cost_vector
df.drop(["Cost"], axis = 1, inplace = True)
df = df[["Supplier Name", "Invoice Number", "Part Number", "new_cost", "Purchase Date"]]
df.rename(columns={'new_cost': 'Cost'}, inplace=True)
print("cost vector average :", df["Cost"].mean())
```

```python
cost_sum = df["Cost"].sum()
cost_avg = df["Cost"].mean()

print("sum :", cost_sum, "average :", cost_avg)
```
