```python
%matplotlib inline

import pandas as pd
import seaborn as sns
```

```python
'''seaborn은 pandas의 dataframe객체와 호환된다.'''

tips = sns.load_dataset("tips")
tips
```

```python
# relplot은 분포도를 그려주는 함수이다. 
# data 매개변수에 dataframe이 들어가는 것이다.

plt.figure(figsize=(5, 5))
sns.relplot(x = "total_bill", y = "tip", data=tips)

```

```python
# hue 매개변수를 사용하면, 특정 범주형 자료에 따른 산포도를 그릴 수 있다.
# style 매개변수도 hue 매개변수와 같은 기능을 수행하지만, style 매개변수는 눈에 잘 띄지 않기 때문에 잘 사용되지 않는다.

sns.relplot(x = "total_bill", y = "tip", data=tips, hue = "size")
```

```python
'''히스토그램을 그리는 방식은 다른 그래프를 그리는 방식과 조금 다르다.'''
# 히스토그램
sns.distplot(tips["total_bill"])
```

```python
# 범주형 간의 히스토그램 비교
sns.distplot(tips["total_bill"][tips["size"] == 2], kde = False)
sns.distplot(tips["total_bill"][tips["size"] == 3], kde = False)
```

```python
# 막대그래프
sns.countplot(x = "time", data = tips)
```
