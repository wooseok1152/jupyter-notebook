```python
import time



sb_driver = webdriver.Chrome(path)
sb_driver.get("https://www.istarbucks.co.kr/store/store_map.do")
time.sleep(2)
sb_driver.find_element_by_class_name("loca_search").click()
time.sleep(2)
sb_driver.find_element_by_class_name("set_sido_cd_btn").click()
time.sleep(2)
sb_driver.find_element_by_class_name("set_gugun_cd_btn").click()
time.sleep(2)
sb = bs4.BeautifulSoup(sb_driver.page_source)
time.sleep(2)
result_list = sb.find_all("p", class_ =  "result_details")
for i in result_list:
    print(i.text)
```
